```python
import matplotlib.pyplot as plt

# ENTER MARKS HERE (example)
marks = [78, 65, 82, 40, 90]

# Calculations
average = sum(marks) / len(marks)
highest = max(marks)
lowest = min(marks)

# Pass / Fail
status = "PASS" if average >= 40 else "FAIL"

print("Marks:", marks)
print("Average:", average)
print("Highest:", highest)
print("Lowest:", lowest)
print("Status:", status)

# Bar chart
subjects = [f"S{i+1}" for i in range(len(marks))]

plt.bar(subjects, marks)
plt.title("Student Marks Analysis")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
